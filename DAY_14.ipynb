# Operations on Vectors

```python
import numpy as np
```

### Dot Product of two vectors

```python
 a = np.array([2,3])

 b = np.array([4,4])

 a_dot_b = np.dot(a,b)
 print(a_dot_b)
```

```python
c = np.array([40,20,35])

d = np.array([53,24,68])

c_dot_d = np.dot(c,d)
print(c_dot_d)
```

### Cross Product of two vectors

```python
a = np.array([2,3])

b = np.array([4,4])

a_cross_b = np.cross(a,b)
print(a_cross_b)
```

```python
c = np.array([5,10,20])

d = np.array([18,32,50])

c_cross_d = np.cross(c,d)
print(c_cross_d)
```

### Projection of "A" vector on "V" vector

```python
a = np.array([2,5])

v = np.array([8,-6])

# magnitude of v
magnitude_of_v = np.sqrt(sum(v ** 2))
# print(magnitude_of_v)
proj_of_a_on_v = (np.dot(a,v)/magnitude_of_v ** 2) * v

print(proj_of_a_on_v)
```

### Create a Matrix using NumPy

### creating matrix with random values

```python
random_matrix = np.random.rand(3,3)
print(random_matrix)
```

```python
# creating random matrix with random integers
random_integer_matrix = np.random.randint(100,size = (4,5))
print(random_integer_matrix)
```

### Matrix with all one's

```python
matrix_3 = np.ones((2,3))
print(matrix_3)
```

```python
matrix_3 = np.ones((2,3),dtype='int')
print(matrix_3)
```

### Null Matrix

```python
null_matrix = np.zeros((4,4))
print(null_matrix)
```

### Identity Matrix

```python
identity_matrix = np.eye(3,3,dtype='int')
print(identity_matrix)
```

### Transpose of a Matrix

```python
# matrix with random integer values
a = np.random.randint(100,size=(4,5))
print(a)
```

```python
transpose_of_a = np.transpose(a)
print(transpose_of_a)
```
