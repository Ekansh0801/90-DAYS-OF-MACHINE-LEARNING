# Matlptolib

```python
# importing matplotlib library

import matplotlib.pyplot as plt
```

```python
# import numpy to get random data for plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

### Plotting the data

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
# adding title to x-axis and y-axis
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
# parabola

x = np.linspace(-10,10,20)
y = x ** 2

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
plt.plot(x,y,'rx')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

### Bar Graph

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','French','Japanese','German','Spanish']
people = [100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel('Languages')
plt.ylabel('Number of People')
plt.show()
```

### Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','French','Japanese','German','Spanish']
people = [100,50,150,40,70]
ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```

### Scatter plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)

fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])

ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')

plt.show()
```

### 3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)

ax.scatter(x,y,z,c=x,cmap='Blues')
plt.show()
```

# Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Note : Seaborn has some built-in datasets

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# visualize the data
sns.set_theme()
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
# load the iris dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
# scatter plot
sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris)
```

```python
sns.scatterplot(x='sepal_length',y='petal_width',hue='species',data=iris)
```

```python
# load the titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
# count plot
sns.countplot(x='class',data=titanic)
```

```python
sns.countplot(x='survived',data=titanic)
```

```python
# Bar Chart

sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```
