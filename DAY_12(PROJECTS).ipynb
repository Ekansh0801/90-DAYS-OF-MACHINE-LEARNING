# ROCK VS MINE PREDICTION

### importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

### data collections and data preprocessing

```python
# loading the data to pandas dataframe
sonar_data = pd.read_csv('/content/sonar data.csv',header=None)
```

```python
sonar_data.head()
```

```python
sonar_data.shape
```

```python
sonar_data.describe()
```

```python
sonar_data[60].value_counts()
```

### M --> Mines

### R ---> Rocks

```python
sonar_data.groupby(60).mean()
```

```python
# seperating data and labels
X = sonar_data.drop(columns=60,axis = 1)
Y = sonar_data[60]
```

```python
print(X)
```

```python
print(Y)
```

### splitting training and testing data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.1, stratify=Y,random_state=1)
```

```python
print(X.shape,X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

### Model Training --->  Logistic regression

```python
model = LogisticRegression()
```

```python
## training the logistic regression model with training data
model.fit(X_train,Y_train)
```

### Model Evaluation

```python
## accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on Training Data itself: ',training_data_accuracy)
```

```python
## accuracy on testing data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on Testing Data: ',testing_data_accuracy)
```

# DIABETES PREDICTION

### Importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

### Data Collection and analysis

```python
# loading the diabetes dataset
diabetes_dataset = pd.read_csv('/content/diabetes.csv')
```

```python
# printing first 5 rows of dataset
diabetes_dataset.head()
```

```python
#printing rows and columns of dataset
diabetes_dataset.shape
```

```python
# printing statistical measures of dataset
diabetes_dataset.describe()
```

```python
# printing number of diabetic and non - diabetic values
diabetes_dataset['Outcome'].value_counts()
```

### 0 ---> Non Diabetic
### 1 ---> Diabetic

```python
# grouping dataset accoring to outcome with mean
diabetes_dataset.groupby('Outcome').mean()
```

```python
## seperating data and label
X = diabetes_dataset.drop(columns='Outcome',axis=1)
Y = diabetes_dataset['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

### Data Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
```

```python
print(standardized_data)
```

```python
X = standardized_data
Y = diabetes_dataset['Outcome']
```

```python
print(X)
print(Y)
```

### Train Test Spilt

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

### Training the model

```python
classifier = svm.SVC(kernel='linear')
```

```python
# training the support vector machine classifer
classifier.fit(X_train,Y_train)
```

### Model Evaluation

#### Accuracy Score

```python
# accuracy score on training model
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy Score of Training Data : ",training_data_accuracy)
```

```python
# accuracy score on testing model
X_test_prediction = classifier.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy Score of Testing Data : ",testing_data_accuracy)
```

### Making a Predictive System

```python
input_data = (6,	148,	72,	35,	0,	33.6,	0.627,	50)

# changing the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the numpy array
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardized the input data
std_data = scaler.transform(input_data_reshaped)
print(std_data)

# predicting
prediction = classifier.predict(std_data)
if prediction[0] == 1:
  print('Person is Diabetic')
else:
  print('Person is non - diabetic')
```

# Spam Mail Prediction

### importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

### Data collection and preprocessing

```python
# loading the data to pandas dataframe
raw_mail_data = pd.read_csv('/content/mail_data.csv')
```

```python
# printing first 5 rows
raw_mail_data.head()
```

```python
## checking for null values
raw_mail_data.isnull().sum()
```

```python
# checking number of rows and columns
raw_mail_data.shape
```

### Label Encoding

```python
# label spam mail as 0 and ham mail as 1

raw_mail_data.loc[raw_mail_data['Category'] == 'spam', 'Category',] = 0
raw_mail_data.loc[raw_mail_data['Category'] == 'ham', 'Category',] = 1
```

```python
raw_mail_data.head()
```

```python
## seperating the data and label
X = raw_mail_data['Message']
Y = raw_mail_data['Category']
```

```python
print(X)
print(Y)
```

### Splitting the data into training and testing data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=3)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

### Feature Extraction

```python
# transform the text data to feature vectors that can be used as input to logisitic regression
feature_extraction = TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

# convert y values as integers
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train_features)
```

```python
print(X_test_features)
```

### Training the model

```python
model = LogisticRegression()
```

```python
# training the logistic regression model with the training data
model.fit(X_train_features,Y_train)
```

### Evaluating the Trained model

```python
# prediction on training data

prediction_on_training_data = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train,prediction_on_training_data)
```

```python
print('Accuracy on Training Data : ',accuracy_on_training_data)
```

```python
# prediction on testing data

prediction_on_testing_data = model.predict(X_test_features)
accuracy_on_testing_data = accuracy_score(Y_test,prediction_on_testing_data)
```

```python
print('Accuracy on Testing Data : ',accuracy_on_testing_data)
```

### Building a Predictive System

```python
input_mail = ["Free entry in 2 a wkly comp to win FA Cup final tkts 21st May 2005. Text FA to 87121 to receive entry question(std txt rate)T&C's apply 08452810075over18's"]

# convert input mail to feature vectors
input_data_features = feature_extraction.transform(input_mail)

# making a prediction
prediction = model.predict(input_data_features)
print(prediction)

if prediction[0] == 1:
  print('Ham Mail')
else:
  print('Spam Mail')
```
